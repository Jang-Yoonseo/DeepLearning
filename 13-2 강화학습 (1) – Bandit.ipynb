```python
class Bandit: #Action을 받으면 보상을 반환
  def step(self,agent_action):
    reward = 1 if agent_action == 0 else 10
    return reward
```

#### 최종 Agent

상황.. 위의 방식을 사용하여 Agent를 완성하라\
t < 0 일 땐 균등하게 action, t >= 20부터 q_table을 바탕으로 한 갱신된 확률로 수행

```python
class Agent:
  def __init__(self):
    self.action_space = [0,1]
    self.action = None
    self.reward = None
    self.actions = []
    self.rewards = []
    self.q_table = np.array([0.001,0.001])
    self.n_experience = 0
  def act(self):
    if self.n_experience <= 20:
      self.action = np.random.choice(self.action_space) #학습 전엔 uniform
    else:
      prob = self.q_table/ self.q_table.sum()
      self.action = np.random.choice(self.action_space, p = prob) #그냥 하나 주면 알아서 첫 사건에 prob 나머지 사건에 1-prob 할당
  def save_experience(self):
    self.actions.append(self.action)
    self.rewards.append(self.reward)
    self.n_experience = self.n_experience + 1
  def learn(self):
    if self.n_experience < 20:
      pass
    else:
      actions = np.array(self.actions)
      rewards = np.array(self.rewards)
      q0,q1 = rewards[actions == 0].mean(), rewards[actions ==1].mean()
      self.q_table = np.array([q0,q1])
```

최종 코드

```python
class Agent:
  def __init__(self):
      self.action_space = [0,1]
      self.action = None
      self.reward = None
      self.actions = []
      self.rewards = []
      self.q_table = np.array([0.001,0.001])
      self.n_experience = 0
  def act(self):
      if self.n_experience <= 20:
        self.action = np.random.choice(self.action_space)
      else:
        prob = self.q_table/ self.q_table.sum()
        self.action = np.random.choice(self.action_space,p = prob )
  def save_experience(self):
    self.actions.append(self.action)
    self.rewards.append(self.reward)
    self.n_experience = self.n_experience + 1
  def learn(self):
    if self.n_experience < 20:
        pass
    else:
        actions = np.array(self.actions)
        rewards = np.array(self.rewards)
        q0,q1 = rewards[actions==0].mean(), rewards[actions==1].mean()
        self.q_table = np.array([q0,q1])
```

```python
env = Bandit()
agent = Agent()

for t in range(1,51):
  agent.act()
  agent.reward = env.step(agent.action)
  agent.save_experience()
  agent.learn()

  print(
      f"t:{t}\t"
      f"action:{agent.action}\t"
      f"reward:{agent.reward}\t"
      f"recent mean:{np.mean(agent.rewards[-20:])}\t"
  )

  if t < 20:
    pass
  elif t == 20:
    print('---')
  else:
    if np.mean(agent.rewards[-20:]) > 9.5:
      print('game clear')
      break
```
